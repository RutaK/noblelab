```python
from openpyxl import load_workbook

import numpy as np

import csv
from scipy import stats
import pandas as pd
import seaborn as sns
import plotly.express as px

import matplotlib.pyplot as plt
import matplotlib

hf = pd.read_excel(r"C:\Users\eart0477\Pycharm projects\data_process\csvfiles\From the lab\December 2020 ophiolites\overnights.xlsx", header=0, index_col=0, sheet_name='He overnights')
nf = pd.read_excel(r"C:\Users\eart0477\Pycharm projects\data_process\csvfiles\From the lab\December 2020 ophiolites\overnights.xlsx", header=0, index_col=0, sheet_name='Ne overnights')
af = pd.read_excel(r"C:\Users\eart0477\Pycharm projects\data_process\csvfiles\From the lab\December 2020 ophiolites\overnights.xlsx", header=0, index_col=0, sheet_name='Ar overnights')
kxf = pd.read_excel(r"C:\Users\eart0477\Pycharm projects\data_process\csvfiles\From the lab\December 2020 ophiolites\overnights.xlsx", header=0, index_col=0, sheet_name='Kr Xe overnights')
hf['Date'] = hf.index
nf['Date'] = nf.index
af['Date'] = af.index
kxf['Date'] = kxf.index
pd.to_datetime(nf['Date'], errors='coerce')
pd.to_numeric(nf['20Ne (H2)'], errors='coerce')
nf = nf.dropna(axis = 0, subset=[ 'Date', '20Ne (H2)'])
nf['20Ne (H2)'] = nf['20Ne (H2)'].astype(float)

```

```python
##this is data clean up step

hhf = nf.loc[nf['20Ne (H2)'] >7000]
hhf = hhf.loc[hhf['20Ne (H2)'] <40000]

hhf = hhf.loc[hhf['20Ne/22Ne'] >8]
hhf = hhf.loc[hhf['20Ne/22Ne'] <13]
hhf = hhf.loc[hhf['21Ne/22Ne'] >0.015]
hhf = hhf.loc[hhf['21Ne/22Ne'] <0.035]
```

```python
#initial plot
fig = px.scatter(nf, x=nf['Date'], y=nf['20Ne (H2)'], log_y=False)


fig.show()
```

```python
#calcs for extra ratios
hhf['20Ne_(H2)_rolling_avg' ] = hhf['20Ne (H2)'].rolling(7).mean()
hhf['20Ne_(H2)_rolling_std' ] = hhf['20Ne (H2)'].rolling(7).std()
hhf['20Ne_(H2)_relative_error'] = hhf['20Ne_(H2)_rolling_std' ]/hhf['20Ne_(H2)_rolling_avg' ]*100

hhf['20Ne/22Ne_rolling_avg' ] = hhf['20Ne/22Ne'].rolling(7).mean()
hhf['20Ne/22Ne_rolling_std' ] = hhf['20Ne/22Ne'].rolling(7).std()

hhf['21Ne/22Ne_rolling_avg' ] = hhf['21Ne/22Ne'].rolling(7).mean()
hhf['21Ne/22Ne_rolling_std' ] = hhf['21Ne/22Ne'].rolling(7).std()

```

```python
#now I'm just plotting the filtered data that was saved separately after previous steps

hhf = pd.read_excel(r"C:\Users\eart0477\Pycharm projects\data_process\csvfiles\From the lab\December 2020 ophiolites\overnight_standards.xlsx", header=0, index_col=0, sheet_name='Ne overnights')
#pd.to_datetime(hhf['Date'], errors='coerce')
```

```python
fig = px.scatter(hhf, x=hhf['Date'], y=hhf['20Ne (H2)'], log_y=False)
fig.add_scatter(name='Rolling average (7 measurments)',x=hhf['Date'], y=hhf['20Ne_(H2)_rolling_avg'], mode='lines')
fig.update_layout(
    title="20Ne (H2 cup) overnight standards",  xaxis_title="Run date", yaxis_title="20Ne fA")
fig.update_layout(legend=dict(yanchor="bottom",y=-0.2,xanchor="left",x=0.01))
fig.update(layout_yaxis_range=[0,40000])

fig.show()
```

```python
fig1 = px.scatter(hhf, x=hhf['Date'], y=hhf['20Ne/22Ne'], log_y=False)
fig1.add_scatter(name='Rolling average (7 measurments)',x=hhf['Date'], y=hhf['20Ne/22Ne_rolling_avg'], mode='lines')
fig1.update_layout(
    title="20Ne/22Ne overnight standards",  xaxis_title="Run date", yaxis_title="20Ne/22Ne")
fig1.update_layout(legend=dict(yanchor="bottom",y=-0.2,xanchor="left",x=0.01))
fig1.update(layout_yaxis_range=[4,12])
fig1.show()
```

```python
fig2 = px.scatter(hhf, x=hhf['Date'], y=hhf['21Ne/22Ne'], log_y=False)
fig2.add_scatter(name='Rolling average (7 measurments)',x=hhf['Date'], y=hhf['21Ne/22Ne_rolling_avg'], mode='lines')
fig2.update_layout(
    title="21Ne/22Ne overnight standards",  xaxis_title="Run date", yaxis_title="21Ne/22Ne")
fig2.update_layout(legend=dict(yanchor="bottom",y=-0.2,xanchor="left",x=0.01))
fig2.update(layout_yaxis_range=[0.01, 0.05])
fig2.show()
```

```python
def print_into_excel(dtp):


        book = load_workbook(r"C:\Users\eart0477\Pycharm projects\data_process\csvfiles\From the lab\December 2020 ophiolites\test2.xlsx")
        writer = pd.ExcelWriter(r"C:\Users\eart0477\Pycharm projects\data_process\csvfiles\From the lab\December 2020 ophiolites\test2.xlsx", engine='openpyxl')
        writer.book = book
        writer.sheets = {ws.title: ws for ws in book.worksheets}

        for sheetname in writer.sheets:
            dtp.to_excel(writer, sheet_name=sheetname, startrow=writer.sheets[sheetname].max_row, index=True,
                         header=True)
        writer.save()

        return

```
