```python
conc1 = '132Xe'
conc2 = '84Kr'
ratio1= '82Kr/84Kr'
ratio2= '130Xe/132Xe'
```

```python
from openpyxl import load_workbook

import numpy as np

import csv
from scipy import stats
import pandas as pd
import seaborn as sns
import plotly.express as px

import matplotlib.pyplot as plt
import matplotlib

hf = pd.read_excel(r"C:\Users\eart0477\Pycharm projects\data_process\csvfiles\From the lab\December 2020 ophiolites\overnights.xlsx", header=0, index_col=0, sheet_name='He overnights')
nf = pd.read_excel(r"C:\Users\eart0477\Pycharm projects\data_process\csvfiles\From the lab\December 2020 ophiolites\overnights.xlsx", header=0, index_col=0, sheet_name='Ne overnights')
af = pd.read_excel(r"C:\Users\eart0477\Pycharm projects\data_process\csvfiles\From the lab\December 2020 ophiolites\overnights.xlsx", header=0, index_col=0, sheet_name='Ar overnights')
kxf = pd.read_excel(r"C:\Users\eart0477\Pycharm projects\data_process\csvfiles\From the lab\December 2020 ophiolites\overnights.xlsx", header=0, index_col=0, sheet_name='Kr Xe overnights')
hf['Date'] = hf.index
nf['Date'] = nf.index
af['Date'] = af.index
kxf['Date'] = kxf.index
pd.to_datetime(kxf['Date'], errors='coerce')
pd.to_numeric(kxf[conc1], errors='coerce')
pd.to_numeric(kxf[conc2], errors='coerce')
kxf = kxf.dropna(axis = 0, subset=[ 'Date', conc1, conc2])
kxf[conc1] = kxf[conc1].astype(float)

```

```python
#initial plot
fig = px.scatter(nf, x=kxf['Date'], y=kxf[conc2], log_y=False)


fig.show()
```

```python

##this is data clean up step

hhf = kxf.loc[kxf[conc1] >150]
hhf = hhf.loc[hhf[conc1] <1017]
hhf = kxf.loc[kxf[conc2] >1903]
hhf = hhf.loc[hhf[conc2] <19323]

hhf = hhf.loc[hhf[ratio1] >0.2]
#hhf = hhf.loc[hhf[ratio1] <400]
hhf = hhf.loc[hhf[ratio2] >0.12]
hhf = hhf.loc[hhf[ratio2] <0.18]
```

```python
#calcs for extra ratios
hhf[f'{conc1}_rolling_avg' ] = hhf[conc1].rolling(7).mean()
hhf[f'{conc1}_rolling_std' ] = hhf[conc1].rolling(7).std()
hhf[f'{conc1}_relative_error'] = hhf[f'{conc1}_rolling_std' ]/hhf[f'{conc1}_rolling_avg'  ]*100

hhf[f'{conc2}_rolling_avg' ] = hhf[conc2].rolling(7).mean()
hhf[f'{conc2}_rolling_std' ] = hhf[conc2].rolling(7).std()
hhf[f'{conc2}_relative_error'] = hhf[f'{conc2}_rolling_std' ]/hhf[f'{conc2}_rolling_avg'  ]*100

hhf[f'{ratio1}_rolling_avg' ] = hhf[ratio1].rolling(7).mean()
hhf[f'{ratio1}_rolling_std' ] = hhf[ratio1].rolling(7).std()

hhf[f'{ratio2}_rolling_avg' ] = hhf[ratio2].rolling(7).mean()
hhf[f'{ratio2}_rolling_std' ] = hhf[ratio2].rolling(7).std()

```

```python
#now I'm just plotting the filtered data that was saved separately after previous steps

#hhf = pd.read_excel(r"C:\Users\eart0477\Pycharm projects\data_process\csvfiles\From the lab\December 2020 ophiolites\overnight_standards.xlsx", header=0, index_col=0, sheet_name='Ne overnights')
#pd.to_datetime(hhf['Date'], errors='coerce')
```

```python
fig = px.scatter(hhf, x=hhf['Date'], y=hhf[conc1], log_y=False)
fig.add_scatter(name='Rolling average (7 measurments)',x=hhf['Date'], y=hhf[f'{conc1}_rolling_avg'], mode='lines')
fig.update_layout(
    title=f"{conc1} overnight standards",  xaxis_title="Run date", yaxis_title=conc1)
fig.update_layout(legend=dict(yanchor="bottom",y=-0.2,xanchor="left",x=0.01))
fig.update(layout_yaxis_range=[0,2000])

fig.show()
```

```python
fig4 = px.scatter(hhf, x=hhf['Date'], y=hhf[conc2], log_y=False)
fig4.add_scatter(name='Rolling average (7 measurments)',x=hhf['Date'], y=hhf[f'{conc2}_rolling_avg'], mode='lines')
fig4.update_layout(
    title=f"{conc2} overnight standards",  xaxis_title="Run date", yaxis_title=conc2)
fig4.update_layout(legend=dict(yanchor="bottom",y=-0.2,xanchor="left",x=0.01))
fig4.update(layout_yaxis_range=[0,30000])

fig4.show()
```

```python
fig = px.scatter(hhf, x=hhf['Date'], y=hhf[f'{conc1}_relative_error'], log_y=False)
fig.show()
```

```python
fig1 = px.scatter(hhf, x=hhf['Date'], y=hhf[ratio1], log_y=False)
fig1.add_scatter(name='Rolling average (7 measurments)',x=hhf['Date'], y=hhf[f'{ratio1}_rolling_avg'], mode='lines')
fig1.update_layout(
    title=f"{ratio1} overnight standards",  xaxis_title="Run date", yaxis_title=ratio1)
fig1.update_layout(legend=dict(yanchor="bottom",y=-0.2,xanchor="left",x=0.01))
fig1.update(layout_yaxis_range=[0.18,0.25])
fig1.show()
```

```python
fig2 = px.scatter(hhf, x=hhf['Date'], y=hhf[ratio2], log_y=False)
fig2.add_scatter(name='Rolling average (7 measurments)',x=hhf['Date'], y=hhf[f'{ratio2}_rolling_avg'], mode='lines')
fig2.update_layout(
    title=f"{ratio2} overnight standards",  xaxis_title="Run date", yaxis_title=ratio1)
fig2.update_layout(legend=dict(yanchor="bottom",y=-0.2,xanchor="left",x=0.01))
fig2.update(layout_yaxis_range=[0.12,0.18])
fig2.show()

```

```python
def print_into_excel(dtp):


        book = load_workbook(r"C:\Users\eart0477\Pycharm projects\data_process\csvfiles\From the lab\December 2020 ophiolites\test2.xlsx")
        writer = pd.ExcelWriter(r"C:\Users\eart0477\Pycharm projects\data_process\csvfiles\From the lab\December 2020 ophiolites\test2.xlsx", engine='openpyxl')
        writer.book = book
        writer.sheets = {ws.title: ws for ws in book.worksheets}

        for sheetname in writer.sheets:
            dtp.to_excel(writer, sheet_name=sheetname, startrow=writer.sheets[sheetname].max_row, index=True,
                         header=True)
        writer.save()

        return

```
