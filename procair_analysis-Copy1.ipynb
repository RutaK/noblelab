```python
from openpyxl import load_workbook

import numpy as np

import csv
from scipy import stats
import pandas as pd
import seaborn as sns
import plotly.express as px

import matplotlib.pyplot as plt
import matplotlib

df = pd.read_excel(r"C:\Users\eart0477\Pycharm projects\data_process\csvfiles\From the lab\December 2020 ophiolites\proc_airs_quad_autosaved.xlsx", header=0, index_col=0, sheet_name='sorted')
df['oilrig_volume'] = df['oilrig_volume'].astype(str)
df['he_mani'] = df['he_mani'].astype(str)
```

```python
def print_into_excel(dtp):


        book = load_workbook(r"C:\Users\eart0477\Pycharm projects\data_process\csvfiles\From the lab\December 2020 ophiolites\test.xlsx")
        writer = pd.ExcelWriter(r"C:\Users\eart0477\Pycharm projects\data_process\csvfiles\From the lab\December 2020 ophiolites\test.xlsx", engine='openpyxl')
        writer.book = book
        writer.sheets = {ws.title: ws for ws in book.worksheets}

        for sheetname in writer.sheets:
            dtp.to_excel(writer, sheet_name=sheetname, startrow=writer.sheets[sheetname].max_row, index=True,
                         header=True)
        writer.save()

        return

    
```

```python

f= df.loc[df['oilrig_volume'].isin(['pip'])]
f= f.loc[f['he_mani'] == 'T']
f = f.dropna(axis = 0, subset=['Baratron', '4He (fA)', 'Date'])
f['4He (fA)'] = f['4He (fA)'].astype(float)
f['fa/mbar'] = f['4He (fA)']/f['Baratron']

f = f.loc[f['fa/mbar'] <14.5]
f = f.loc[f['fa/mbar'] > 2]


f['rolling_avg' ] = f['fa/mbar'].rolling(7).mean()
f['rolling_std' ] = f['fa/mbar'].rolling(7).std()

#print_into_excel(f)
```

```python
fig = px.scatter(f, x=f['Date'], y=f['fa/mbar'], log_y=False)
fig.add_scatter(name='Pippette',x=f['Date'], y=f['rolling_avg'], mode='lines')
fig.add_scatter(name='upper bound',x=f['Date'], y=f['rolling_avg']+f['rolling_std'], mode='lines',  
                marker=dict(color="#444"),line=dict(width=0), showlegend=False)
fig.add_scatter(name='lower bound',x=f['Date'], y=f['rolling_avg']-f['rolling_std'], mode='lines',
                marker=dict(color="#444"),line=dict(width=0), showlegend=False, fillcolor='rgba(68, 68, 68, 0.15)',
        fill='tonexty')
      
      



fig.update_layout(
    title="4He signal (fA)/mbar for fixed analytical volumes",
    xaxis_title="Time(s)",
    yaxis_title="m torr")
fig.update_xaxes

fig.show()

#f['rolling_avg'],f['rolling_std' ]
```
